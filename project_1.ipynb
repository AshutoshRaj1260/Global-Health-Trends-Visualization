**Now, taking out the average Life Expectancy of each Year**

```python
l=[]
for i in range(2000, 2016):
  df= lifeExp_df[lifeExp_df['Year']==i]
  l.append(df['Life expectancy '].mean())

l
```

##Analyzing the change in the Life Expectancy year by year

```python
# Create a line chart with labels
y = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]
plt.figure(figsize=(7,4))
plt.plot(l, y, label='Data Line', marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Life Expectancy')
plt.ylabel('Year')
plt.title('Line Chart Example')

plt.legend()

plt.grid(True)

plt.show()
```

##Next, categorizing the Countries on the basis of their Development Status and looking at the average Life Expectancy in the respective nations

```python
print("Development Status of Countries in the list: ")
lifeExp_df.Status.nunique()
```

```python
# Category 1 consists of the developing nations

category_1_df = lifeExp_df[lifeExp_df['Status']=='Developing']
category_1_df.head()
```

****

```python
# Category 2 consists of the developed nations

category_2_df = lifeExp_df[lifeExp_df['Status']=='Developed']
category_2_df.head()
```

```python
dev1_avg = category_1_df['Life expectancy '].mean()
dev1_avg
```

```python
dev2_avg = category_2_df['Life expectancy '].mean()
dev2_avg
```

```python
p1 = dev1_avg / (dev1_avg + dev2_avg) * 100
p2 = dev2_avg / (dev1_avg + dev2_avg) * 100

labels = ['Developing Nation', 'Developed Nations']
sizes = [p1, p2]

colors = ['#ff9999', '#66b3ff']

explode = (0.1, 0)
plt.figure(figsize=(8,4))
plt.pie(sizes, autopct=lambda p: '{:.1f}'.format(p*(dev1_avg+dev2_avg)/100), startangle=90, colors=colors, explode=explode, shadow=True)

plt.axis('equal')

plt.title('Distribution of Development')

plt.legend(labels, loc='upper right')

plt.show()
```
